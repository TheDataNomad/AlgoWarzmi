# Pandas and Numpy

<ol>
   <li> The data manipulation capabilities of pandas are built on top of the numpy library. In a way, numpy is a dependency of the pandas library. </li>
    
   <li>Pandas is best at handling tabular data sets comprising different variable types (integer, float, double, etc.) In addition, the pandas library can also be used to perform even the most naive of tasks such as loading data or doing feature engineering on time series data. </li>
   <li> Numpy is most suitable for performing basic numerical computations such as mean, median, range, etc. Alongside, it also supports the creation of multi-dimensional arrays. </li>
    
   <li> Remember, python is a zero indexing language unlike R where indexing starts at one.</li>
   <li> The best part of learning pandas and numpy is the strong active community support you'll get from around the world. </li> 
    </ol>

# Numpy

```python
import numpy as np
```

```python
#The difference between numpy arrays and normal list arrays - They are homogonoes 

lstOne = [1,2,3,4,5,6,7,8,9]
print(type(lstOne))
print(lstOne)

lstTwo = np.array(lstOne,dtype=str)
[print(type(lstTwo))]
print(lstTwo)
```

```python
#creating arrays
np.zeros(10, dtype='int')


#creating a 3 row x 5 column matrix
np.ones((3,5), dtype=float)


#creating a matrix with a predefined value
np.full((3,5), 10)

#create an array with a set sequence
np.arange(0, 20, 2)

#create an array of even space between the given range of values
np.linspace(0, 1, 5)
```

```python
#slicing and indexing is exactly the same as list data strucutre 
```

```python
# Array Concatenation

#You can concatenate two or more arrays at once.
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
z = [21,21,21]
np.concatenate([x, y,z])
```

```python
#You can also use this function to create 2-dimensional arrays.
grid = np.array([[1,2,3],[4,5,6]])
np.concatenate([grid,grid])


#Using its axis parameter, you can define row-wise or column-wise matrix
np.concatenate([grid,grid],axis=1)
```
