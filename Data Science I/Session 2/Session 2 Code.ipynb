```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv", index_col=0)

data.head()
```

## ScatterPlot

### Pandas

```python
data.plot(kind = 'scatter',x = 'Attack',y = 'Defense')
```

### Matplotlib

```python
plt.scatter("Attack","Defense",data=data)
```

### Seaborn

```python
sns.scatterplot("Attack","Defense",data=data)
```

```python
# Plotting Layouts

f,ax = plt.subplots(1,2,figsize=(18,8))
data.plot(kind = 'scatter',x = 'Attack', y = 'Defense',ax = ax[0], label = 'Pokemon Attack/Defense pair')
ax[0].set_title('Attack Vs Defense')


sns.lineplot(data=data['HP'], ax=ax[1])
ax[1].set_title('HP Levels across all Pokemons')

plt.show()
```
