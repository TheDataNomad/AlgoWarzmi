```python
g = sns.FacetGrid(data,col="class")
g = g.map(sns.boxplot,"sepal width")
```

```python
g = sns.FacetGrid(data,col="class",xlim= Aut, ylim=None)
g = g.map(sns.histplot,"sepal length")
```

```python
g = sns.PairGrid(data)
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.regplot)
```
